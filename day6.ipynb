```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
```

```python
x=np.array([1,5,13,25,33,46]).reshape((-1,1))
y=np.array([3,6,15,12,21,18])
```

```python
plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
SLR=LinearRegression()
```

```python
SLR.fit(x,y)
```

```python
pred=SLR.predict(x)
```

```python
from sklearn.metrics import mean_squared_error


mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='Red',marker='*')

print("MSE:", mse)
print("intercept:", SLR.intercept_)
print("slope:", SLR.coef_)
```

```python
import random
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


random.seed(42)
num_points = 50
x = [random.uniform(0, 10) for _ in range(num_points)]
y = [2 * xi + 3 + random.gauss(0, 1) for xi in x]  # y = 2x + 3 + noise


x = np.array(x).reshape(-1, 1)
y = np.array(y)


model = LinearRegression()


model.fit(x, y)


y_pred = model.predict(x)


plt.scatter(x, y, label='Original data')
plt.plot(x, y_pred, color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression')
plt.show()


print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Read the Excel file

data = pd.read_excel('/content/drive/MyDrive/Values.xlsx')

# Assuming your Excel columns are named 'x' and 'y'
x = data['x'].values.reshape(-1, 1)
y = data['y'].values

# Create a linear regression model
model = LinearRegression()
model.fit(x, y)

# Predict y values using the model
y_pred = model.predict(x)

# Plot the original data and the regression line
plt.scatter(x, y, label='Original Data')
plt.plot(x, y_pred, color='red', label='Linear Regression')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Linear Regression')
plt.legend()
plt.show()

# Print the slope (coefficients) and intercept of the regression line
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
