```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y=4+ 2* np.sin(2*x)
#plot
fig, a = plt.subplots()

a.plot(x,y,linewidth=1.0)

a.set(xlim=(0, 8), xticks=np.arange(1, 8),
      ylim=(0, 8), yticks=np.arange(1, 8))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
Temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120])
Res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608])

plt.plot(Temp,Res)
plt.grid()
plt.show()
plt.title("Thermistor characteristics")
plt.xlabel("Temp")
plt.ylabel("Res")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["Karnataka","Ap","Tn","Tg"])
y=np.array([65,68,89,86])

plt.bar(x,y)
plt.show()

plt.pie(y)
plt.show()

plt.plot(x,y)
plt.grid()
plt.show()
```

Histograms - a graphical representation which shows data is distributed

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.random.normal(5, 1, 100000)

plt.hist(x,1000)
plt.show()
```

```python
import random
import numpy
import matplotlib.pyplot as plt

x=numpy.random.randint(0,100)

plt.hist(x,10)
plt.show()
```
