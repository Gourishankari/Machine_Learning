***3.The student should have attendance greater than 75% and marks 45 then they are eligible or else print not eligible***

```python
att = float(input("Enter the student's attendance percentage: "))
mks = float(input("Enter the student's marks: "))

if att > 75 and mks > 45:
    print("The student is eligible.")
else:
    print("The student is not eligible.")
```

**5.The user has to select a number whether it is positive or negative or zero and should also print the type**

```python
num = int(input("Enter a number: "))

if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
print(type(num))
```
